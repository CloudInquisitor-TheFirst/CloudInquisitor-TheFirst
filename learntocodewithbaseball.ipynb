## Lists

Lists are built with square brackets and are basically a simple way to hold other, ordered pieces of data.

    roster_list = ['clayton kershaw', 'mookie betts', 'cody bellinger']

Every spot in a list has a number associate with it.  The first spot is 0. You can get sections (called slices) of your list by separating numbers with a colon.  Both single numbers and slices are called inside square brackets, i.e. [].

```python
roster_list = ['clayton kershaw', 'mookie betts', 'cody bellinger']
roster_list[0]
```

```python
roster_list[0:2]
```

Passing a slice a negative number gives you the end of the list. To get the last two items of a list, you could do:

```python
roster_list[-2:]
```

Also note how when you leave off the number after the colon the slice will automaticallys use the end of the list.

Lists can hold anything, including other lists. Lists that hold other lists are often called nested lists.

## Dicts
A dict is short for dictionary. You can think about it like an actual dictionary if you want. Real dictionaries have words and definitions, Python dicts have **keys** and **values**.

Dicts are basically a way to hold data and give each piece a name.  Dicts are written with curly brackets, like this:

```python
roster_dict = {'P': 'zac gallen',
               'RF': 'corbin carroll',
               'CF': 'alek thomas'}
```

You can access items in a dict like this:

```python
roster_dict['RF']
```

And add new things to dict like this:

```python
roster_dict['1B'] = 'josh naylor'

roster_dict
```

Notice how keys are strings (they're surrounded in quotes).  THey can also be numbers or even bools.  THey cannot be a variable that has not already been created. You could do this:

```python
pos = 'RF'

roster_dict[pos]
```

Because when you run it Python is just replacing pos with 'RF'
But you will get an error if pos is undefined. You also get an error if you try to access a key that's not present in the dict(note : assigning something to a key that isn't ther yet - like we did with 'josh naylor' above - is OK).
While dictionary keys are usually strings, dictionary values can be anything, including lists or other dicts.

## Unpacking

    p, rf = ['zac galledn', 'corbin carroll']

that does the exact same thing as assigning these separately on their own line.

    p = 'merrill kelly'
